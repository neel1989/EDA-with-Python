```python
df.drop_duplicates(inplace=True)
```

```python
df.isna().sum()
```

```python
sns.heatmap(df.isnull())
```

```python
df.mode().iloc[0]
```

```python
df.fillna(df.mode().iloc[0], inplace=True)
```
